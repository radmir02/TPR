## Задание 2

Дана переменная, в которой хранится четырехзначное число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы:
1.  

```
year = 2020
```
Результат:  
`Високосный год` 

2.  

```
year = 2019
```
Результат:  
`Обычный год` 

````python
year_input = int(input('Введите год '))
if year_input % 4 == 0 and year_input % 100 != 0 or year_input % 400 == 0:
    print ("Год високосный ")
else :
    print("Год не високосный ")
````

```
width = 10
length = 205
height = 5
```

Используйте следующие правила:
- если каждое из трех измерений менее 15 сантиметров, то выведите на экран "Коробка №1";
- если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите "Коробка №2";
- если длина товара больше 2 метров, то выводите "Упаковка для лыж";
- во всех остальных случаях выводите "Стандартная коробка №3".


````python
width = 10
length = 205
height = 5
if width < 15 and length < 15 and height < 15:
           print('Коробка №1')
elif length > 200:
           print('Упаковка для лыж')
elif 15 < width < 50 or 15 < length < 50 or 15 < height < 50:
           print('Коробка №2')
else:
           print('Стандартная коробка №3')
````

1. ```number = 123456```

Результат:  
`Неасчастливый билет`
2. ```number = 123321```

`Счастливый билет`


````python
number = 123456
number1 = 123321
num = list (str(number))
num1 = list (str(number1))
res1 = (int(num[0]) + int(num[1]) + int (num[2]))
res11 = (int(num[3]) + int(num[4]) + int (num[5]))
res2 = (int(num1[0]) + int(num1[1]) + int (num1[2]))
res22 = (int(num1[3]) + int(num1[4]) + int (num1[5]))
if res1 == res11:
    print('Билет №1 счастливый')
else :
     print('Билет №1 несчастливый')
if res2 == res22:
    print('Билет №2 счастливый')
else :
    print('Билет №2 несчастливый')
````
